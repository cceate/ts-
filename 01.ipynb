```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('csv/income.csv')
```

```python
data
```

```python
plt.scatter(data.Education,data.Income)
```

# 构造出 f(x) = ax + b 即求出参数a, b 

```python
x=data.Education
y=data.Income
```

# 初始化顺序模型

```python
model=tf.keras.Sequential()
```

# 建立模型

```python
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

# 编译模型

```python
model.summary()
```

# 训练模型

```python
model.compile(optimizer='adam',loss='mse')
```

# 训练模型

```python
model.fit(x,y,epochs=1000)
```

# 查看预测值现有的x进行预测

```python
model.predict(x)
```

# 预测一个未知的edu对应的money值

```python
model.predict(pd.Series([30]))
```
